```python
import numpy as np
import pandas as pd
import statistics as stats
```

```python
# Create NumPy array
arr = np.array([10, 20, 30, 40, 50])

print("Array:", arr)
print("Mean:", np.mean(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# Create DataFrame
data = {
    'Name': ['Aadi', 'CHETAN', 'UTSAV', 'AKSHAY'],
    'Marks': [85, 90, 78, 88]
}

df = pd.DataFrame(data)
df
```

```python
print("Average Marks:", df['Marks'].mean())
print("Highest Marks:", df['Marks'].max())
print("Lowest Marks:", df['Marks'].min())
```

```python
marks = df['Marks'].tolist()

print("Mean:", stats.mean(marks))
print("Median:", stats.median(marks))
print("Mode:", stats.mode(marks))
```

```python
import numpy as np

b = np.array([
    [10, 20, 30, 50, 10],
    [1, 2, 3, 5, 6],
    [1, 1, 1, 1, 1]
])

print(b)
print("Dim:", b.ndim)
print("Shape:", b.shape)
print("Size:", b.size)
print("Item Size:", b.itemsize)
print("Data type:", b.dtype)
```
